# Task 1: Exploring and Visualizing a Simple Dataset 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Task 1: Load the dataset
print("=== LOADING THE DATASET ===")
```

```python
# Task 1: Load the datase
print("=== LOADING THE DATASET ===")
# Load iris dataset from seaborn
iris = sns.load_dataset('iris')
print("Dataset loaded successfully!\n")
```

```python
# Task 2: Display dataset structure
print("=== DATASET STRUCTURE ===")
print("Dataset Shape:", iris.shape)
```

```python
print("\nDataset Columns:")
print(iris.columns.tolist())
```

```python
print("\nData Types:")
print(iris.dtypes)
```

```python
print("\nFirst 5 rows:")
print(iris.head())
```

```python
print("\nBasic Statistics:")
print(iris.describe())
```

```python
print("\nClass distribution:")
print(iris['species'].value_counts())
```

```python
# Task 3: Create basic visualizations

# Scatter Plots to analyze relationships between variables
print("\n=== SCATTER PLOTS ===")
plt.figure(figsize=(15, 10))

# Scatter plot 1: Sepal length vs Sepal width
plt.subplot(2, 2, 1)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', s=100, alpha=0.7)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Scatter plot 2: Petal length vs Petal width
plt.subplot(2, 2, 2)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', s=100, alpha=0.7)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# Scatter plot 3: Sepal length vs Petal length
plt.subplot(2, 2, 3)
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', s=100, alpha=0.7)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')

# Scatter plot 4: Sepal width vs Petal width
plt.subplot(2, 2, 4)
sns.scatterplot(data=iris, x='sepal_width', y='petal_width', hue='species', s=100, alpha=0.7)
plt.title('Sepal Width vs Petal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')

plt.tight_layout()
plt.show()
```

```python
# Histograms to examine data distribution
print("\n=== HISTOGRAMS ===")
plt.figure(figsize=(15, 10))

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold']

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    plt.hist(iris[feature], bins=15, color=colors[i], alpha=0.7, edgecolor='black')
    plt.title(f'Distribution of {feature.replace("_", " ").title()}')
    plt.xlabel(f'{feature.replace("_", " ").title()} (cm)')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Histograms by species
print("Creating Histograms by Species...")
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    for species in iris['species'].unique():
        species_data = iris[iris['species'] == species][feature]
        plt.hist(species_data, alpha=0.6, label=species, bins=10)
    plt.title(f'Distribution of {feature.replace("_", " ").title()} by Species')
    plt.xlabel(f'{feature.replace("_", " ").title()} (cm)')
    plt.ylabel('Frequency')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Box Plots to detect outliers and spread of values
print("\n=== BOX PLOTS ===")
plt.figure(figsize=(15, 10))

# Individual box plots
plt.subplot(2, 2, 1)
sns.boxplot(data=iris, y='sepal_length')
plt.title('Box Plot - Sepal Length')

plt.subplot(2, 2, 2)
sns.boxplot(data=iris, y='sepal_width')
plt.title('Box Plot - Sepal Width')

plt.subplot(2, 2, 3)
sns.boxplot(data=iris, y='petal_length')
plt.title('Box Plot - Petal Length')

plt.subplot(2, 2, 4)
sns.boxplot(data=iris, y='petal_width')
plt.title('Box Plot - Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Box plots by species
print("Creating Box Plots by Species...")
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=iris, x='species', y=feature)
    plt.title(f'{feature.replace("_", " ").title()} by Species')
    plt.xlabel('Species')
    plt.ylabel(f'{feature.replace("_", " ").title()} (cm)')

plt.tight_layout()
plt.show()
```

```python
# Additional summary statistics
print("\n=== ADDITIONAL ANALYSIS ===")
print("Correlation Matrix:")
numeric_iris = iris.select_dtypes(include=[np.number])
correlation_matrix = numeric_iris.corr()
print(correlation_matrix)
```

```python
# Heatmap of correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```

```python
# Check for missing values
print("\nMissing Values:")
print(iris.isnull().sum())

print("\n=== ANALYSIS COMPLETE ===")
print("Key Insights:")
print("- The dataset contains 150 samples with 4 features and 1 target variable")
print("- Three species: setosa, versicolor, virginica (50 samples each)")
print("- Petal measurements show clearer separation between species")
print("- Some potential outliers visible in sepal width")
print("- Strong correlation between petal length and petal width")
```
